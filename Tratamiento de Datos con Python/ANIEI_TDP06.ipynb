```python
import pandas as pd
import numpy as np

import random
from datetime import datetime, timedelta, date
```

```python
df = pd.read_json('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/alumnos_calif_jsonArray.json')
df.info()
```

```python
df
```

```python
#Ver cuantos nulos o faltantes existen en cada columna
df.isnull()
```

```python
df.isnull().sum()
```

```python
#Visualizar los rengloes con sexo y curp NaN
df[pd.isnull(df['sexo']) & pd.isnull(df['curp'])]
```

```python
#Obtener los indices de renglon de aquellos que tienen valor Nan en calificacion
df_na_calif = df[pd.isnull(df['calificacion'])]['calificacion'].isna()
df_na_calif
```

```python
df['calificacion'].describe()
```

```python
#Visualizamos los renglones con faltantes en calificacion
df[df.index.isin(df_na_calif.index)]
```

```python
#Cambiar los NaN en calificacion
# df.fillna(value='') cambiar todos los valores faltantes del df

#Cambiar los vales faltantes de una sola columna del df
#df['calificacion'].fillna(value=0, inplace=True)
df['calificacion'].fillna((df['calificacion'].min()), inplace=True)
```

```python
#Visualizamos los renglones con faltantes en calificacion
df[df.index.isin(df_na_calif.index)]
```

```python
df.isnull().sum()
```

```python
#Agregar columna
#Calcular GPA (Grade Point Average)
df['gpa'] = (((df['calificacion'] * 100) / df['calificacion'].max()) / 20) -1
```

```python
#Agregar una columna fecha basa en un serie de tiempo generada con un rango
sti = pd.date_range("2018-01-01", periods=df.clave_alu.count(), freq='d')
print(df.clave_alu.count())
print(sti)
df['fecha'] = sti
```

```python
df
```

```python
#Listar calificaciones mayores a 5 ordenados de forma descendente por sexo y calificacion
df[df['calificacion'] > 5].sort_values(by=['sexo','calificacion'], ascending=False)
```

```python
#Eliminar columnas
df.drop(['curp'], axis=1, inplace=True)
df
```

```python
df.info()
```

```python
#Convertir columnas de tipo object a tipo fecha
df['fecha'] = pd.to_datetime(df['fecha'])

#Convertir columnas de tipo object a tipo str usando apply()
df['ap_paterno'] = df['ap_paterno'].apply(str)

#Convertir columnas de tipo object a tipo str usando astype()
df['clave_alu'] = df['clave_alu'].astype(str)
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df['fecha'].describe()) #datetime_is_numeric=True
print("Primer fecha", df['fecha'].min())
print("Ultima fecha", df['fecha'].max())
```

```python
#Filtrar renglones por fecha
df[df['fecha'] >= datetime(year=2020, month=1, day=1)].sort_values(by='fecha')
```

```python
#Filtrar renglones por fecha y sexo
df.loc[df['fecha'] >= datetime(year=2020, month=1, day=1)].sort_values(by='fecha')
```

```python
df_dup = df.duplicated(subset=['ap_paterno','ap_materno','nombre','sexo','ciudad','email'], keep='last')
df_dup
```

```python
#Visualizar los duplicados
df[df_dup]
```

```python
#Eliminar duplicados
df.drop_duplicates(subset=['ap_paterno','ap_materno','nombre','sexo','ciudad','email'], 
                   keep='last',
                   inplace=True)
df.count()
```

```python
df
```

```python
#Visualizar los valores unicos de ciudad
df['ciudad'].unique()
```

```python
#Visualizar el numero de apariciones por valor en ciudad
df['ciudad'].value_counts()
```

```python
#Reemplazar datos no validos usando un diccionario, la llave es el valor a cambiar y el valor es eñ nuevo dato
dic_queretaro = {"QUERETRO":"QUERETARO", "QUER":"QUERETARO", "QUERETARO QRO.":"QUERETARO", 
                 "SANTIAGO DE QUERETARO":"QUERETARO", "QUERETAROQRO":"QUERETARO", "QRO.":"QUERETARO"}

df['ciudad'] = df['ciudad'].replace(dic_queretaro)
df['ciudad'].value_counts()
```

```python
#Cambiar valores a mayusculas
df['ap_paterno'] = df['ap_paterno'].str.upper()
df['ap_materno'] = df['ap_materno'].str.upper()
df['nombre'] = df['nombre'].str.upper()
df
```

```python
#Generar un columna escalar
gpa_rango = [1, 2, 3, 4, 5]
gpa_etiqueta = ['D', 'C', 'B', 'A']
df['gpa_escala'] = pd.cut(df['gpa'], bins=gpa_rango, labels=gpa_etiqueta)
df.head(20)
```
